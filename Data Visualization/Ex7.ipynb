```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
bnb_filepath = "../input/new-york-city-airbnb-open-data/AB_NYC_2019.csv"
```

```python
bnb_data = pd.read_csv(bnb_filepath)
bnb40_data = bnb_data[:40]
```

```python
bnb_data.head()
```

```python
plt.figure(figsize=(24,6))
sns.scatterplot(x=bnb40_data['neighbourhood'], y=bnb40_data['price'])
```

```python
sns.scatterplot(x=bnb40_data['reviews_per_month'], y=bnb40_data['price'])
```

```python
sns.regplot(x=bnb40_data['reviews_per_month'], y=bnb40_data['price'])
```

```python
sns.scatterplot(x=bnb40_data['reviews_per_month'], y=bnb40_data['price'], hue=bnb40_data['room_type'])
```

```python
sns.lmplot(x='reviews_per_month', y='price', hue='room_type', data=bnb40_data)
```

```python
sns.swarmplot(x=bnb40_data['room_type'],
              y=bnb40_data['price'])
```
